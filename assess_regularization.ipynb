```python
import pickle
import torch
with open('observations/train_scores_b64lr0.0003d3w32', 'rb') as f:
    train_scores = pickle.load(f)
with open('observations/test_scores_b64lr0.0003d3w32', 'rb') as f:
    test_scores = pickle.load(f)

train_scores = torch.tensor(train_scores)
test_scores = torch.tensor(test_scores)
```

```python
import matplotlib.pyplot as plt
plt.plot(train_scores)
plt.show()
```

```python
plt.plot(test_scores)
plt.show()
```

```python
medians_ts = [chunk.median() for chunk in torch.split(test_scores, 100)]
medians_tr = [chunk.median() for chunk in torch.split(train_scores, 587)]
```

```python
plt.plot(medians_ts, label="test")
plt.plot(medians_tr, label="train")
plt.legend()
plt.show()
```
